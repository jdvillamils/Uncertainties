```python
#Run once
import sympy as sp
import numpy as np
import math
```

```python
# Define interest variables at left and right
u=sp.Symbol('u')
a=sp.Symbol('a')
c=sp.Symbol('c')

 



#Add variables to the next list
var=[u,a,c,]

#Define equation to compute uncertainties
Ec=((u*np.square(3.14159265359))/(2*a*c))**2

#--
derivates=[]
for element in var:
  d=sp.diff(Ec,element)
  derivates.append(d)
size=len(var)
NumDer=[]
#--


#Introduce numerical values in |subs={...}|, first the variable then the value  
for j in range(size):
  nd=derivates[j].evalf(subs={u:1.04148e-6,a:4.60e-4,c:0.155})#HERE
  NumDer.append(nd)


#Define Values of each uncertainty in the same order as the variables
inc=[0,3e-6,0.0001]


ITotal=0
for h in range(size):
  ITotal=ITotal+(abs(NumDer[h]))*inc[h]


print("Uncertainty is", ITotal, "Bye")
  

  

```
