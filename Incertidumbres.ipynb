```python
#corre primero esta celda (solo una vez)
import sympy as sp
import numpy as np
import math
```

```python
# Define las variables a usar en la izquierda y derecha 
u=sp.Symbol('u')
a=sp.Symbol('a')
c=sp.Symbol('c')

 



#Agrégalas en esta lista, separadas por comas 
var=[u,a,c,]

#Define la Ecuación a calcularle la incertidumbre (para elevar se usa **, i.e. x al cuadrado sería x**2)
Ec=((u*np.square(3.14159265359))/(2*a*c))**2

#--No hay que hacer nada aquí :v
derivates=[]
for element in var:
  d=sp.diff(Ec,element)
  derivates.append(d)
size=len(var)
NumDer=[]
#--


#Define los valores numéricos de cada variable en la seccion subs={...}, primero la variable y luego de los : el valor numerico
for j in range(size):
  nd=derivates[j].evalf(subs={u:1.04148e-6,a:4.60e-4,c:0.155})#AQUI
  NumDer.append(nd)


#Define el valor de las incertidumbres en el orden en que se definieron las variables, seguidos por comas
inc=[0,3e-6,0.0001]


ITotal=0
for h in range(size):
  ITotal=ITotal+(abs(NumDer[h]))*inc[h]


print("La incertidumbre total es", ITotal, "Bye")
  

  

```
